```python
# Package import
import ibis
import os

# Set use case info
use_case_db = 'sr_grids'
use_case_path = '/data/discovery/sr_grids'

# Set hdfs and impalad host info
hdfs_namenode = 'da74wbname02.opr.statefarm.org'
impalad_host = 'da74wbdn01.opr.statefarm.org'

# Set ibis temp db for impala and temp path for HDFS.  This is done to prevent
# a permessions error when ibis attempts to create a temp db.  Set to your use
# case info
ibis.options.impala.temp_db = use_case_db
ibis.options.impala.temp_hdfs_path = use_case_path

# Get/set webhdfs port
hdfs_port = os.environ.get('IBIS_WEBHDFS_PORT', 50070)

# Connect to webhdfs
hdfs = ibis.hdfs_connect(host=hdfs_namenode, port=hdfs_port, auth_mechanism='GSSAPI', verify=False)

# Connect to impala
con = ibis.impala.connect(host=impalad_host, database=use_case_db,
                          hdfs_client=hdfs, auth_mechanism='GSSAPI')
```

```python
# Example usage of connection

# Instantiate table object from use case
table = con.table('zcta')

# Set expression and execute the query
expr = table.limit(10)
query = expr.execute(async=True)

# Print the result
query.get_result()
```
